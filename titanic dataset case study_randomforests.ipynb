```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("train.csv")
```

```python
le=LabelEncoder()
```

```python
df.Sex=le.fit_transform(df.Sex)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.Age=df.loc[:,['Age']].fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,[2,4,5,6,7]]
len(x)
```

```python
y=df.iloc[:,1]
len(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
print(len(xtrain),len(xtest),len(ytrain),len(ytest))
```

```python
r=RandomForestClassifier(n_estimators=100)
```

```python
r.fit(xtrain,ytrain)
```

```python
ypredict=r.predict(xtest)
```

```python
accuracy_score(ypredict,ytest)
```
