```python
df.drop(columns='EmpNumber',inplace=True) # Remvoving Employee number which is not of any value
```

```python
# converting all categorical features to numerical

l=LabelEncoder()
```

```python
df.Gender=l.fit_transform(df.Gender)
df.EducationBackground=l.fit_transform(df.EducationBackground)
df.MaritalStatus=l.fit_transform(df.MaritalStatus)
df.EmpDepartment=l.fit_transform(df.EmpDepartment)
df.EmpJobRole=l.fit_transform(df.EmpJobRole)
df.BusinessTravelFrequency=l.fit_transform(df.BusinessTravelFrequency)
df.Attrition=l.fit_transform(df.Attrition)
df.OverTime=l.fit_transform(df.OverTime)
```

```python
# Extracting the best features from available predictor variables

from sklearn.feature_selection import SelectKBest,chi2
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,26]
```

```python
sel=SelectKBest(chi2,k=3)    # selected 12 important features by using SelectKBest
```

```python
sel.fit(x,y)
```

```python
sel.transform(x)
```

```python
# Spliting the data into train and test

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
r=RandomForestClassifier()
```

```python
r.fit(xtrain,ytrain)
```

```python
ypredict=r.predict(xtest)
```

```python
accuracy_score(ypredict,ytest)
```

```python
confusion_report=confusion_matrix(ypredict,ytest)
```
