```python
#####Placeholder concept
a=tf.placeholder(tf.float32)
b=tf.placeholder(tf.float32)
c=a*b
print(c)
```

```python
#sess=tf.Session()
#print(sess.run(c,{a:[0,4,6],b:[9,6,7]}))
#sess.close()
```

```python
with tf.Session() as sess:
  output=sess.run(c,{a:[5,6,6,5],b:[6,8,4,9]})
  print(output)
```

```python
#varible concept...

##simple Linear concept

## y =mx+c
import tensorflow as tf
```

```python
m=tf.Variable(0.3,tf.float32)
c=tf.Variable(0.5,tf.float32)
x=tf.placeholder(tf.float32)
y=m*x+c
init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
print(sess.run(y,{x:[1,5,7,9,3]}))
```
